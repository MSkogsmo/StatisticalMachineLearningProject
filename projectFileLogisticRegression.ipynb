```python
#Divide set into training and testing randomly, sampling indices for training
# Size of test set
size = 300

trainI = np.random.choice(training_data.shape[0], size, replace=False)
trainIndex = training_data.index.isin(trainI)
train = training_data.iloc[trainIndex] # training set
test = training_data.iloc[~trainIndex] # test set
```

```python
#  
model = skl_lm.LogisticRegression()
X_train = train[['acousticness','danceability','duration',
                'energy','instrumentalness','key','liveness',
                'loudness','mode','speechiness','tempo','time_signature',
                'valence','label']]
Y_train = train['label']
X_test = test[['acousticness','danceability','duration',
                'energy','instrumentalness','key','liveness',
                'loudness','mode','speechiness','tempo','time_signature',
                'valence','label']]
Y_test = test['label']
model.fit(X_train, Y_train)
print('Model summary:')
print(model)
```

```python
predict_prob = model.predict_proba(X_test)
print('The class order in the model:')
print(model.classes_)
print('Examples of predicted probablities for the above classes:')
predict_prob[0:5] # inspect the first 5 predictions
```

```python
prediction = np.empty(len(X_test), dtype=object)
prediction = np.where(predict_prob[:, 0]>=0.5, 0, 1)
prediction[0:100] # Inspect the first 10 predictions after labeling.
```

```python
# Confusion matrix
print(pd.crosstab(prediction, Y_test))
# Accuracy
np.mean(prediction == Y_test)
```

```python
false_positive_rate = []
true_postive_rate = []
N = np.sum(Y_test == 0)
P = np.sum(Y_test == 1)
threshold = np.linspace(0.01, 0.99, 99)
model = skl_lm.LogisticRegression(solver='liblinear')
model.fit(X_test, Y_test)
predict_prob = model.predict_proba(X_test)

for i in range(len(threshold)):
    prediction = np.empty(len(X_test), dtype=object)
    prediction = np.where(predict_prob[:, 0] > threshold[i], 0, 1)
    FP = np.sum((prediction==1)&(Y_test==0))
    TP = np.sum((prediction==1)&(Y_test==1))
    false_positive_rate.append(FP/N)
    true_postive_rate.append(TP/P)
```

```python
plt.plot(false_positive_rate,true_postive_rate)
plt.show()
```
